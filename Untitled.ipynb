```python
X = data[['Lat', 'Lon']]
# Instanciate DBSCAN 
db = DBSCAN(eps=.005, min_samples=5, metric="manhattan")
db.fit(X)
clusters = db.labels_
# Number of clusters created
print(np.unique(clusters))
```

```python
X['cluster'] = [str(i) for i in clusters]
big_cluster = X["cluster"].value_counts().index[0:6].tolist()
X["cluster"].value_counts()
```

```python
dff = X
cluster_to_keep = [i for i in big_cluster if i != "-1"]

mapbox_access_token = 'pk.eyJ1IjoibGV3aXN3ZXJuZWNrIiwiYSI6ImNsMnMzYnA1OTA5dXgza25yazhhajh3NGsifQ.Z3CrpQqY0Xj_ZH3spiAiYQ'

fig = px.scatter_mapbox(
    dff[dff["cluster"].isin(cluster_to_keep)],
    lat="Lat",
    lon="Lon",
    color="cluster",
    size_max=15,
    zoom=10,
    height=600
)

fig.update_layout(
    template="plotly_dark", 
    mapbox_style="dark",
    mapbox=dict(accesstoken=mapbox_access_token),
    title=f"{len(cluster_to_keep)} clusters found on day of week {dayofweek} and hour {hour}",
    title_x=.5
)

fig.show()
```
