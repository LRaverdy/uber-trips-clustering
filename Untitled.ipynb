```python
import pandas as pd
import numpy as np
import datetime as dt
import plotly.express as px

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.cluster import AgglomerativeClustering
```

```python
df = pd.read_csv("uber-raw-data-jul14.csv")

display(df.head())

print(f"Data shape :{df.shape}")
```

#### Change date format

```python
df["Date/Time"] = pd.to_datetime(df["Date/Time"])
```

```python
df['hour'] = df['Date/Time'].dt.hour
df['dayofweek'] = df['Date/Time'].dt.dayofweek
df['dayofmonth'] = df['Date/Time'].dt.day
```

```python
data = df.groupby("dayofmonth")["dayofmonth"].agg(["count"]).reset_index().sort_values(by="count", ascending=False)
data
```

```python
fig = px.bar(
    data, 
    x="dayofmonth", 
    y="count", 
    text="count", 
    color="count", 
    color_continuous_scale=["red",  "darkblue", "lightblue"]
)

fig.update_layout(
    template="plotly_dark", 
    height=400,
    width=1000,
    title="Total pick ups by day of month July 14", 
    title_x=.5
)

fig.update_traces(textposition="outside")
fig.update_traces(texttemplate='%{text:.2s}')

fig.show("png")
```

```python
data = df.groupby("hour")["hour"].agg(["count"]).reset_index()
data["hour"] = data["hour"].astype(str)

fig = px.bar(
    data, 
    x="hour", 
    y="count", 
    text="count", 
    color="count", 
    color_continuous_scale=["red",  "darkblue", "lightblue"]
)

fig.update_layout(
    template="plotly_dark",
    height=400,
    width=800,
    title="Total pick ups by hour on july 14", 
    title_x=.5
)

fig.update_traces(texttemplate='%{text:.2s}')

fig.show("png")
```

```python
data = df.groupby("dayofweek")["dayofweek"].agg(["count"]).reset_index()

data["dayofweek"] = data["dayofweek"].apply(lambda x: "Monday" if x == 0
                                           else "Tuesday" if x == 1
                                           else "Wednesday" if x == 2
                                           else "Thursday" if x == 3
                                           else "Friday" if x == 4
                                           else "Saturday" if x == 5
                                           else "Sunday")

fig = px.bar(
    data, 
    x="dayofweek", 
    y="count", 
    text="count", 
    color="count",
    color_continuous_scale=["red",  "darkblue", "blue", "lightblue"]
)

fig.update_layout(
    template="plotly_dark", 
    title="Total pick ups by day of week on july 14", 
    title_x=.5, 
    height=300, 
    width=800
)

fig.update_traces(texttemplate='%{text:.2s}')

fig.show("png")
```

We are able to notice a few things from these 3 graphs :
- 15th of July has the most important count of pick ups in 24h
- Hours between 15 and 22 are the most frequented hours while hours early in the morning are poorly frequented.
- Most of the trips occur on Wednesday and Thurday

But now we need to have a deeper view on what is going on. Let's make a heatmap !

```python
matrix = df.groupby(["dayofweek", "hour"])["hour"].agg(["count"]).reset_index().sort_values(by=["hour", "dayofweek"])
matrix["hour"] = matrix["hour"].astype(str)
matrix["dayofweek"] = matrix["dayofweek"].astype(str)


data = [matrix[matrix["dayofweek"] == str(i)]["count"] for i in range(0,7)]
```

```python
matrix["dayofweek"] = matrix["dayofweek"].apply(lambda x: "Monday" if x == "0"
                                           else "Tuesday" if x == "1"
                                           else "Wednesday" if x == "2"
                                           else "Thursday" if x == "3"
                                           else "Friday" if x == "4"
                                           else "Saturday" if x == "5"
                                           else "Sunday")
```

```python
fig = px.imshow(
    data,
    labels=dict(x="Hour", y="Day of week", color="Total"),
    x=matrix["hour"].unique(),
    y=matrix["dayofweek"].unique(),
    color_continuous_scale=["red",  "darkblue", "blue", "lightblue"]
)

fig.update_layout(
    template="plotly_dark",
    title="Pick ups by day and hour",
    height=400,
    width=1000,
    title_x=.5,
    legend_title=None,
    legend_orientation='h',
    legend_valign='middle',
    legend_x=-0.01,
    legend_y=-0.1
)

fig.show("png")
```

#### The dataset contains almost 800.000 rows, we're going to take a sample before applying tuned model on whole data

```python
dff = df.sample(50000)
```

```python
dayofweek = 2
hour = 21

day = '2014-07-17'
#data = df[(df['Date/Time'].dt.strftime('%Y-%m-%d') == day)]
data = dff[dff["dayofweek"] == dayofweek]
data = data[data["hour"] == hour]
X = data[['Lat', 'Lon']]
```

```python
#X = data[['Lat', 'Lon']]
# Instanciate DBSCAN 
db = DBSCAN(eps=.004, min_samples=6, metric="manhattan")
db.fit(X)
clusters = db.labels_
# Number of clusters created
print(np.unique(clusters))
```

```python
X['cluster'] = [str(i) for i in clusters]
big_cluster = X["cluster"].value_counts().index[0:4].tolist()
X["cluster"].value_counts()
```

```python
dff = X
cluster_to_keep = [i for i in big_cluster if i != "-1"]

mapbox_access_token = 'pk.eyJ1IjoibGV3aXN3ZXJuZWNrIiwiYSI6ImNsMnMzYnA1OTA5dXgza25yazhhajh3NGsifQ.Z3CrpQqY0Xj_ZH3spiAiYQ'

fig = px.scatter_mapbox(
    dff[dff["cluster"].isin(cluster_to_keep)],
    lat="Lat",
    lon="Lon",
    color="cluster",
    size_max=15,
    zoom=10,
    height=500
)

fig.update_layout(
    template="plotly_dark", 
    mapbox_style="dark",
    mapbox=dict(accesstoken=mapbox_access_token),
    title=f"{len(cluster_to_keep)} decent clusters generally form on day of week {dayofweek} and hour {hour}",
    title_x=.5
)

fig.show("png")
```
